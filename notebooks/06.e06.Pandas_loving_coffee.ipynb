```python
renamed_df = df[['Country.of.Origin', 'Producer', 'Processing.Method']].copy()
renamed_df.rename(columns={
    'Processing.Method': 'Processing method',
    'Country.of.Origin': 'Country of origin'},
                  inplace=True)
```

```python
def generate_plot(df, column):   
    data = [
        go.Bar(
            x=df[column].value_counts().index,
            y=df[column].value_counts().values
        )
    ]

    fig = go.Figure(data=data)
    
    fig.update_layout(
        title_text=column,
        title_x=0.5,
        title_font_size=42,
        font_size=18)
    
    fig.update_layout(
        xaxis_title= column,
        yaxis_title="Number of counts",
    )
    fig.update_layout(template='plotly_dark')
    fig.show()
```

```python
for col in renamed_df.columns:
    generate_plot(renamed_df, col)
```
